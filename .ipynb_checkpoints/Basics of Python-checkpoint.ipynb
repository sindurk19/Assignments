```python
# Exercise 1: Prime Numbers

def is_prime(number):
    """Checks if a given number is prime.

    Args:
        number: The number to check.

    Returns:
        True if the number is prime, False otherwise.
    """
    if number <= 1:
        return False
    for i in range(2, int(number**0.5) + 1):
        if number % i == 0:
            return False
    return True

# Example usage
num = 17
if is_prime(num):
    print(f"{num} is a prime number.")
else:
    print(f"{num} is not a prime number.")
```

```python
# Exercise 2: Product of Random Numbers

import random

def check_product():
    """Generates two random numbers, asks for their product, and checks the answer."""
    num1 = random.randint(1, 10)  # Generates numbers between 1 and 10 (inclusive)
    num2 = random.randint(1, 10)
    product = num1 * num2

    try:
        user_answer = int(input(f"What is the product of {num1} and {num2}? "))
        if user_answer == product:
            print("Correct!")
        else:
            print(f"Incorrect. The correct answer is {product}.")
    except ValueError:
        print("Invalid input. Please enter an integer.")

# Example usage
check_product()
```

```python
# Exercise 3: Squares of Even/Odd Numbers

def print_squares(start, end, even_or_odd="even"):  # Default to even
    """Prints the squares of even or odd numbers within a range.

    Args:
        start: The starting number of the range.
        end: The ending number of the range.
        even_or_odd: "even" or "odd" to specify which numbers to square.
    """
    print(f"Squares of {even_or_odd} numbers between {start} and {end}:")

    for num in range(start, end + 1):
        if even_or_odd == "even" and num % 2 == 0:
            print(f"{num}^2 = {num**2}")
        elif even_or_odd == "odd" and num % 2 != 0:
            print(f"{num}^2 = {num**2}")


# Example Usage (Even numbers)
print_squares(100, 200, "even")

# Example Usage (Odd numbers)
print_squares(100, 200, "odd")
```

```python
# Exercise 4: Word Counter

def word_counter(input_text):
    """Counts the number of occurrences of each word in a text.

    Args:
        input_text: The text to analyze.

    Returns:
        A dictionary where keys are words and values are their counts.
    """
    word_counts = {}
    words = input_text.split()  # Split the text into words

    for word in words:
        # Remove punctuation from word (optional, but good practice):
        word = word.strip('.,!?').lower() #make it case insensitive and strip punctuation
        if word in word_counts:
            word_counts[word] += 1
        else:
            word_counts[word] = 1
    return word_counts

# Example usage
input_text = "This is a sample text. This text will be used to demonstrate the word counter."
counts = word_counter(input_text)
for word, count in counts.items():
    print(f"'{word}': {count}")
```

```python
# Exercise 5: Check for Palindrome

def is_palindrome(text):
    """Checks if a string is a palindrome.

    Args:
        text: The string to check.

    Returns:
        True if the string is a palindrome, False otherwise.
    """
    processed_text = ''.join(char.lower() for char in text if char.isalnum()) # Remove non-alphanumeric and make it lowercase
    return processed_text == processed_text[::-1]  # Compare with reversed string


# Example usage
print(is_palindrome("racecar"))  # Output: True
```
