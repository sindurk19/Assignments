```python
#Step 1: Import Libraries & Load Dataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.ensemble import RandomForestClassifier, BaggingClassifier, AdaBoostClassifier, GradientBoostingClassifier

# Load dataset
data = pd.read_csv("C:\\Users\\Sindu\\Desktop\\Assignment ExcelR\\Random Forest\\Random Forest\\glass.csv")

# Show first few rows
print(data.head())
print("\nShape:", data.shape)
 
```

```python
#Step 2: Exploratory Data Analysis (EDA)
# Dataset info
print(data.info())

# Check missing values
print("\nMissing values:\n", data.isnull().sum())

# Basic statistics
print("\nDescriptive Statistics:\n", data.describe())
```

```python
#Step 3: Data Visualization
# Histograms
data.hist(figsize=(12,10), bins=20)
plt.show()

# Boxplots to detect outliers
plt.figure(figsize=(12,8))
sns.boxplot(data=data)
plt.xticks(rotation=45)
plt.show()

# Correlation heatmap
plt.figure(figsize=(10,8))
sns.heatmap(data.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.show()

# Distribution of target variable
sns.countplot(x="Type", data=data, palette="Set2")
plt.title("Class Distribution")
plt.show()
```

```python
#Step 4: Data Preprocessing
# Features and Target
X = data.drop("Type", axis=1)
y = data["Type"]

# Check imbalance
print("Class distribution:\n", y.value_counts())

# Feature scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Step 5: Random Forest Model Implementation

# Clean target variable
y_clean = y.dropna()
X_clean = X.loc[y_clean.index]   # keep only rows where target is not missing

# Rescale features again for aligned X
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled_clean = scaler.fit_transform(X_clean)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled_clean, y_clean, test_size=0.2, random_state=42, stratify=y_clean
)

# Random Forest Classifier
rf = RandomForestClassifier(
    n_estimators=200,
    random_state=42,
    class_weight="balanced"
)

# Train
rf.fit(X_train, y_train)

# Predict
y_pred = rf.predict(X_test)

# Evaluation
print("✅ Random Forest Results")
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
#Step 6: Bagging vs Boosting
# Bagging with Decision Tree
bagging = BaggingClassifier(n_estimators=200, random_state=42)
bagging.fit(X_train, y_train)
y_bag = bagging.predict(X_test)

print("Bagging Accuracy:", accuracy_score(y_test, y_bag))

# AdaBoost
adaboost = AdaBoostClassifier(n_estimators=200, random_state=42)
adaboost.fit(X_train, y_train)
y_ada = adaboost.predict(X_test)

print("AdaBoost Accuracy:", accuracy_score(y_test, y_ada))

# Gradient Boosting
gb = GradientBoostingClassifier(n_estimators=200, random_state=42)
gb.fit(X_train, y_train)
y_gb = gb.predict(X_test)

print("Gradient Boosting Accuracy:", accuracy_score(y_test, y_gb))
```

```python
#Step 7: Explanation (Theory)

✅ Bagging (Bootstrap Aggregating):

Trains multiple models independently on different bootstrapped samples.

Example: Random Forest = Bagging + Decision Trees.

Reduces variance → avoids overfitting.

✅ Boosting:

Trains models sequentially, each new model focuses on mistakes made by the previous one.

Examples: AdaBoost, Gradient Boosting, XGBoost.

Reduces bias → improves accuracy but more prone to overfitting.
```
