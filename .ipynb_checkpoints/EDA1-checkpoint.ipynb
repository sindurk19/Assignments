```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("C:\\Users\\Sindu\\Desktop\\Assignment ExcelR\\EDA1\\EDA1\\Cardiotocographic.csv")

# 1. Data Cleaning and Preparation

# Rename columns for clarity (optional, but recommended)
df.columns = ['LB', 'AC', 'FM', 'UC', 'DL', 'DS', 'DP', 'ASTV', 'MSTV', 'ALTV', 'MLTV', 'NSP', 'Tendency', 'Width']  # Assuming NSP and CLASS are present

# Check for missing values
print(df.isnull().sum())  # Display missing value counts per column

# Handle missing values (example: imputation with median for numerical features)
for col in ['LB', 'AC', 'FM', 'UC', 'DL', 'DS', 'DP', 'ASTV', 'MSTV', 'ALTV', 'MLTV']:
    df[col].fillna(df[col].median(), inplace=True) # Or df[col].fillna(df[col].mean(), inplace=True) if mean is preferred

# Check data types
print(df.dtypes)

# Convert data types if necessary (e.g., if numerical values are stored as strings)
# Example: df['LB'] = pd.to_numeric(df['LB'], errors='coerce') # Use errors='coerce' to handle non-numeric values gracefully

# Detect and treat outliers (example: using IQR method for 'LB')
Q1 = df['LB'].quantile(0.25)
Q3 = df['LB'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df['LB'] = np.clip(df['LB'], lower_bound, upper_bound) # Or remove outliers entirely: df = df[(df['LB'] >= lower_bound) & (df['LB'] <= upper_bound)]

# Repeat outlier treatment for other relevant columns as needed

# 2. Statistical Summary

print(df.describe())  # Provides descriptive statistics for numerical columns

# 3. Data Visualization

# Histograms for numerical variables
df.hist(figsize=(15, 12)) # Adjust figure size as needed
plt.show()

# Box plots for numerical variables
plt.figure(figsize=(15, 10))
sns.boxplot(data=df) # Or create individual boxplots for specific columns
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels if needed
plt.show()

# Correlation heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()

# Scatter plots (example: LB vs. AC)
plt.figure(figsize=(8, 6))
sns.scatterplot(x='LB', y='AC', data=df)
plt.title("Scatter Plot of LB vs. AC")
plt.show()

# Pair plots for exploring relationships between multiple variables
sns.pairplot(df[['LB', 'AC', 'FM', 'NSP']]) # Select relevant columns
plt.show()

# Violin plots (example: distribution of LB for different NSP classes)
plt.figure(figsize=(8, 6))
sns.violinplot(x='NSP', y='LB', data=df)
plt.title("Violin Plot of LB for Different NSP Classes")
plt.show()

# 4. Pattern Recognition and Insights (Examples)

# Correlation analysis (from the heatmap)
correlation_lb_ac = df['LB'].corr(df['AC'])
print(f"Correlation between LB and AC: {correlation_lb_ac}")

# Analyze distributions and potential relationships (from histograms, box plots, scatter plots)
# Example: Look for patterns in LB distribution for different NSP classes (normal, suspect, pathological)

# 5. Conclusion (In the Jupyter Notebook, summarize your findings)

# Example: "The EDA revealed a negative correlation between LB and AC, suggesting..."
# "The distribution of LB varies across different NSP classes, with..."


# Report (Summarize key findings and insights)
# Example: "The analysis of the cardiotocographic dataset revealed several important patterns..."
# "Further investigation is recommended to..."
```
